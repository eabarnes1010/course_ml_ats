## Choose your image!
* https://scikit-image.org/docs/stable/auto_examples/data/plot_general.html#sphx-glr-auto-examples-data-plot-general-py

```python
## MODIFY: uncomment an image below to use it in this notebook

# image = data.chelsea()[:,:,1]
# image = data.rocket()[:,:,-1]  #<--- highly recommended!
# image = data.camera()
# image = data.checkerboard()
# image = data.stereo_motorcycle()[0][:,:,0]
image = data.brick()

np.shape(image)
```

```python
#----------------------------------------------------------------------------------------
# Don't mess with this code -  smooth the data a bit with a Gaussian kernel 
# to make it nicer to work with
smooth_image = filters.gaussian(image, 1)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image)
plt.title('original image')

plt.subplot(1,2,2)
plt.imshow(smooth_image)
plt.title('slightly smoothed')

plt.show()
#----------------------------------------------------------------------------------------
```

#### Identity Kernel

```python
kernel = np.array([[0, 0, 0],
                   [0, 1, 0],
                   [0, 0, 0]])


plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.heatmap(kernel,cmap=plt.cm.RdBu_r,annot=True,center=0,cbar=False)
plt.yticks([])
plt.xticks([])

plt.subplot(1,3,2)
plt.imshow(smooth_image)
plt.title('original image')

plt.subplot(1,3,3)
kernel_response = convolve(smooth_image, kernel)
plt.imshow(kernel_response)
plt.title('convolved image')

plt.show()
```

#### Averaging Kernel

```python
kernel = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]])

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.title('kernel')
sns.heatmap(kernel,cmap=plt.cm.RdBu_r,annot=True,center=0,cbar=False)
plt.yticks([])
plt.xticks([])

plt.subplot(1,3,2)
plt.imshow(smooth_image)
plt.title('original image')

plt.subplot(1,3,3)
kernel_response = convolve(smooth_image, kernel)
# kernel_response = convolve(kernel_response, kernel)     # uncomment below to smooth even more
# kernel_response = convolve(kernel_response, kernel)     # uncomment below to smooth even more
# kernel_response = convolve(kernel_response, kernel)     # uncomment below to smooth even more
plt.imshow(kernel_response)
plt.title('convolved image')

plt.show()
```
